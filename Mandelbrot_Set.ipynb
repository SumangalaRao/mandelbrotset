```python
"""
Created on April 10,2019
@author:Sumangala Rao
"""
#importing the required libraries
from PIL import Image
from numpy import complex,array
import colorsys
```

```python
#setting the width
wth=1080
```

```python
#returns a tuple of colors which is of the type integer equi for RGB
def convertintorgb(i):
    color = 255*array(colorsys.hsv_to_rgb(i / 255.0, 1, 0.5)) 
    return tuple(color.astype(int))
```

```python
#actual function that implements the mandelbrotset logic
def mandelbrotset(x, y): 
    c0 = complex(x, y) 
    z = 0
    for i in range(1, 1000): 
        if abs(z) > 2: 
            return convertintorgb(i) 
        z = z * z + c0 
    return (0, 0, 0) 
```

```python
#creates a image of the required/specified size here,1080x540
img = Image.new('RGB', (wth, int(wth / 2))) 
pixels = img.load()
```

```python
for x in range(img.size[0]): 
    print("%.2f %%" % (x / wth * 100.0))  
    for y in range(img.size[1]): 
        pixels[x, y] = mandelbrotset((x - (0.75* wth)) / (wth / 4), 
                                      (y - (wth / 4)) / (wth / 4)) 
```
